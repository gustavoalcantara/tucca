Esse arquivo é a tentativa da integração do banco de dados do laboratório (que tem esses nomes 'PLANILHA GERAL ANO') e do REDCAP. Os códigos aqui fazem a leitura e padronização no padrão do REDCAP para subida dos registros.

```python
#Importe as bibliotecas
import pandas as pd
import numpy as np
```

```python
#Inicialmente fiz para 2023 decorrente da solicitação do Fernando.
#geral_2021 = pd.read_excel('2018.xlsx', sheet_name='Plan1')
#geral_2022 = pd.read_excel('2022.xlsx', sheet_name='Plan1')

#Leia as bases de dados
geral_2023 = pd.read_excel('2023.xlsx', sheet_name='Plan1')

#Leia o Dicionário
dicionario = pd.read_csv('dicionario.csv')
```

```python
#Renomeie as variáveis
geral_2023 = geral_2023.rename({'PACIENTE': 'nome',
                                'Data de Nascimento': 'dt_nasc',
                                'Data solicitação (registrada no formulário)':'dt_sol',
                                'DATA ENTRADA TUCCA': 'dt_ent_lab',
                                'MATERIAL RECEBIDO': 'mat_rec',
                                'Tipo de amostra': 'local_amostra',
                                'Diagnóstico Inicial': 'hip_diag',
                                'Serviço de Origem': 'nome_servico',
                                'Médico Solicitante': 'nome_medico',
                                'Contato': 'fone_medico',
                                'MATERIAL PARA REVISÃO?':'revisao_caso',
                                'Data de envio para LABPAC fazer Cortes ou IHQ': 'dt_ent_lab',
                                'DATA DA REVISÃO PELO PATOLOGISTA ': 'dt_revisao',
                                'DATA ENVIO DO LAUDO FINAL DA PATOLOGIA/IMUNO PARA O MEDICO RESPONSÁVEL': 'dt_entrega_laudo'},
                               axis = 'columns')
print(geral_2023.columns)
```

```python
#Selecione as variáveis
#Essas variaveis são as que constam no REDCAP
geral_2023 = geral_2023[['nome', 'dt_nasc', 'dt_sol', 'dt_ent_lab', 'mat_rec', 'local_amostra', 'hip_diag', 'nome_servico', 'nome_medico', 'fone_medico', 'revisao_caso', 'dt_ent_lab', 'dt_revisao', 'dt_entrega_laudo']]
```

```python
#Transforme em dataframe.
dicionario = pd.DataFrame(dicionario)
geral_2023 = pd.DataFrame(geral_2023)
```

```python
#Transponha para virar no formato do REDCAP
novo_df = dicionario['Variable / Field Name']
novo_df = novo_df.to_frame().T.reset_index(drop = True)
novo_df = novo_df.rename(columns=novo_df.iloc[0]).drop(novo_df.index[0])
novo_df
```

```python
#Verifique quais são as variáveis comuns aos dois df.
# Encontre as colunas comuns entre os dois dataframes
common_cols = geral_2023.columns.intersection(novo_df.columns)

# Remova as colunas comuns do segundo dataframe
novo_df = novo_df.drop(columns=common_cols)

# Agora faça o merge
z = pd.concat([geral_2023, novo_df], axis=1)
```

```python
#Trasnforme as variáveis de data como data.
#df['data_column'] = pd.to_datetime(df['data_column'], errors='coerce').dt.date
z['dt_nasc'] = pd.to_datetime(z['dt_nasc'], errors = 'coerce').dt.date
z['data_entrega_laudo'] = pd.to_datetime(z['data_entrega_laudo'], errors = 'coerce').dt.date
z['dt_revisao'] = pd.to_datetime(z['dt_revisao'], errors = 'coerce').dt.date
z['dt_sol'] = pd.to_datetime(z['dt_sol'], errors = 'coerce').dt.date
z['dt_teste'] = pd.to_datetime(z['dt_teste'], errors = 'coerce').dt.date
```

```python
#Dê o Append e atribua ao 'Padrão' do RedCap
#z = x.append(y, ignore_index = True)
z = z.applymap(lambda x: str(x).split('.')[0] if pd.notna(x) and str(x).lower() != 'nan' else "")
```

```python
#Crie o record_id
z['record_id'] = range(1, len(z) + 1)
z = z.set_index('record_id').reset_index()
```

```python
#Verifique as colunas duplicadas
colunas_duplicadas = z.columns[z.columns.duplicated()]
z = z.loc[:,~z.columns.duplicated()]
z['dt_ent_lab'] = pd.to_datetime(z['dt_ent_lab'], errors='coerce')
z['dt_ent_lab']
```

```python
#Exclua as colunas
colunas_para_excluir = ['mat_rec', 'dt_entrega_laudo', 'material_rec', 'material_rec_2', 'local_corpo', 'tecnica', 'fish_qual', 'qpcr_qual', 'dpcr_qual', 'seq_qual', 'imuno_hist']
z = z.drop(columns=colunas_para_excluir)
```

```python
#atribua espaço em branco para essas duas variaveis
z['revisao_caso'] = ''
z['hip_diag'] = ''
```

```python
#Salve no padrão do RedCap
z.to_csv('input.csv', sep = ',', index=False, header=True)
```
